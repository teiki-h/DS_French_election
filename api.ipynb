```python
#import library
import requests
import time
import pandas as pd
```

### Obtention de la liste des stations

```python
# Define headers for the API request, including the API key and the accept header
headers = {
    'accept': '*/*',
    'apikey': api_climat
}
# Define the delay between requests
limit = 45 
delay = 60 / limit 

stations = ['quotidienne', 'horaire', 'infrahoraire-6m']
datas = []
for type_station in stations:
    l = []
    for departement_id in range(1, 96):
        url = f'https://public-api.meteofrance.fr/public/DPClim/v1/liste-stations/{type_station}?id-departement={departement_id}'
        # Make the GET request to the API
        response = requests.get(url, headers=headers)
        df = pd.DataFrame(response.json())
        l.append(df)
        # Print a message indicating which station type and department is being processed
        message = f'type = {type_station}; département = {departement_id}'
        print(message)
        # Wait for the calculated delay before making the next request ( the API limit the number of request per minutes)
        time.sleep(delay)
    df = pd.concat(l)
    # Add a new column to indicate the frequency/type of station
    df['freq'] = type_station
    # Append the final DataFrame to the `datas` list
    datas.append(df)
liste_station_climat = pd.concat(datas)
```
